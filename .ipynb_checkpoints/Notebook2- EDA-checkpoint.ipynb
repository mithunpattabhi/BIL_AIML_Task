```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv("Datasets/data_clean.csv")
data.head()
```

```python
#building scatter plot of total-square feet vs price
plt.figure(figsize=(8,5))
plt.scatter(
    data['total_sqft'],
    data['price'],
    alpha=0.3
)
plt.xlabel("Total Square Feet")
plt.ylabel("Price (in Lakhs)")
plt.title("Total Square Feet vs Price (Filtered)")
plt.show()
```

```python
#Scatter plot - bhk vs price
plt.figure(figsize=(8,5))
sn.scatterplot(x=data['bhk'], y=data['price'])
plt.xlabel("Number of Bedrooms (BHK)")
plt.ylabel("Price (in Lakhs)")
plt.title("BHK vs Price")
plt.show()
```

```python
#histogram: Price distribution
plt.figure(figsize=(8,5))
plt.hist(data['price'], bins=30)
plt.xlabel("Price (in Lakhs)")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```
